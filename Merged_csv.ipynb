```python
import pandas as pd
import glob
import os
import json

folder_path = "/content/drive/MyDrive/GEE_Forest_TimeSeries/"
all_files = glob.glob(os.path.join(folder_path, "*.csv"))

df_list = [pd.read_csv(f) for f in all_files]
merged_df = pd.concat(df_list, ignore_index=True)

print("Columns:", merged_df.columns)

# If .geo exists, extract lat/lon
if '.geo' in merged_df.columns:
    merged_df['longitude'] = merged_df['.geo'].apply(lambda x: json.loads(x)['coordinates'][0])
    merged_df['latitude']  = merged_df['.geo'].apply(lambda x: json.loads(x)['coordinates'][1])

# Sort safely
if {'latitude','longitude','year'}.issubset(merged_df.columns):
    merged_df = merged_df.sort_values(by=['latitude','longitude','year'])

merged_df.to_csv(
    folder_path + "TamilNadu_Forest_TimeSeries_Master.csv",
    index=False
)

print("✅ Master CSV created successfully")
```

```python
# =====================================================
# 🔹 Standard Merged CSV for Odisha Forest Time-Series
# =====================================================

import pandas as pd
import glob
import os
import json

# -----------------------------
# 1️⃣ Folder path for Odisha
# -----------------------------
folder_path = "/content/drive/MyDrive/GEE_Forest_TimeSeries/OD"

# -----------------------------
# 2️⃣ Load all CSV files
# -----------------------------
all_files = glob.glob(os.path.join(folder_path, "*.csv"))

df_list = []
for f in all_files:
    # Skip empty or almost empty files (~2 bytes)
    if os.path.getsize(f) > 10:
        tmp = pd.read_csv(f)
        tmp.columns = tmp.columns.str.strip()  # strip spaces
        df_list.append(tmp)

if len(df_list) == 0:
    raise ValueError(f"No valid CSV files found in {folder_path}!")

merged_df = pd.concat(df_list, ignore_index=True)

print("Columns after merge:", merged_df.columns.tolist())

# -----------------------------
# 3️⃣ Extract latitude/longitude from .geo if needed
# -----------------------------
if '.geo' in merged_df.columns:
    merged_df['longitude'] = merged_df['.geo'].apply(lambda x: json.loads(x)['coordinates'][0])
    merged_df['latitude']  = merged_df['.geo'].apply(lambda x: json.loads(x)['coordinates'][1])

# -----------------------------
# 4️⃣ Keep only essential columns
# -----------------------------
features = ['NDVI','NDMI','NBR','BSI']
required_cols = ['latitude','longitude','year'] + features
required_cols = [c for c in required_cols if c in merged_df.columns]

merged_df = merged_df[required_cols]
merged_df = merged_df.dropna()

# -----------------------------
# 5️⃣ Sort for consistency
# -----------------------------
merged_df = merged_df.sort_values(by=['latitude','longitude','year']).reset_index(drop=True)

print("Merged dataframe shape:", merged_df.shape)
print(merged_df.head())

# -----------------------------
# 6️⃣ Save Master CSV
# -----------------------------
output_csv = folder_path + "_Forest_TimeSeries_Master.csv"
merged_df.to_csv(output_csv, index=False)
print(f"✅ Master CSV created successfully: {output_csv}")
```

```python
import pandas as pd
import glob
import os
import json

folder_path = "/content/drive/MyDrive/GEE_Forest_TimeSeries/PB"
all_files = glob.glob(os.path.join(folder_path, "*.csv"))

df_list = []

for f in all_files:
    try:
        df = pd.read_csv(f)

        # Skip completely empty files
        if df.empty:
            print(f"⚠️ Skipped empty file: {os.path.basename(f)}")
            continue

        df_list.append(df)

    except pd.errors.EmptyDataError:
        print(f"⚠️ Skipped corrupt/blank file: {os.path.basename(f)}")

# Stop if nothing valid
if len(df_list) == 0:
    raise ValueError("❌ No valid CSV files found!")

# Merge
merged_df = pd.concat(df_list, ignore_index=True)

print("Columns:", merged_df.columns)


# =====================================================
# Extract latitude & longitude safely
# =====================================================

if '.geo' in merged_df.columns:

    def extract_lon(x):
        try:
            return json.loads(x)['coordinates'][0]
        except:
            return None

    def extract_lat(x):
        try:
            return json.loads(x)['coordinates'][1]
        except:
            return None

    merged_df['longitude'] = merged_df['.geo'].apply(extract_lon)
    merged_df['latitude']  = merged_df['.geo'].apply(extract_lat)


# =====================================================
# Safe Sorting
# =====================================================

if {'latitude','longitude','year'}.issubset(merged_df.columns):
    merged_df = merged_df.sort_values(by=['latitude','longitude','year'])


# =====================================================
# Save Master File
# =====================================================

output_path = folder_path + "_Forest_TimeSeries_Master.csv"
merged_df.to_csv(output_path, index=False)

print("✅ Master CSV created successfully")
print("📊 Final shape:", merged_df.shape)
```
